```python
import cv2
import matplotlib.pyplot as plt
```

```python
# define essentials
camera_port = 0
ramp_frames = 30
```

```python
# initialize camera object
camera = cv2.VideoCapture(camera_port)
```

```python
# take picture and delete camera object
retval, image = camera.read()
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
del(camera)
```

```python
# show the image
plt.imshow(image)
```
